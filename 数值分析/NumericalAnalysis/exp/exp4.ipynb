```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

%matplotlib inline
```

```python
x = np.array([0,3,5,7,9,11,12,13,14,15])
y = np.array([0,1.2,1.7,2.0,2.1,2.0,1.8,1.2,1.0,1.6])
xnew = np.linspace(0, 15, num=151, endpoint=True)
```

```python
plt.scatter(x, y)
```

```python
def lagrange(x, y):
    from scipy.interpolate  import lagrange
    ret = lagrange(x,y)
    lagy = ret(xnew)
    #作图
    plt.plot(x, y, 'r*')
    plt.plot(xnew, lagy, 'b-')
    plt.title('lagr Interpolation')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
lagrange(x, y)
```

# 很明显使用三次样条

```python
from scipy.interpolate import interp1d
f3 = interp1d(x, y, kind='cubic') # 三次

plt.plot(x, y, 'o', xnew, f3(xnew), '-')
plt.legend(['data', 'cubic'], loc='best')
plt.title("cubic")
plt.show()
```
